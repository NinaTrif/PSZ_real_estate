# Linear Regression custom

In this jupyter notebook, we will be using self implemented Linear Regression model, which we compare to previously evaluated Linear Regression model from sklearn library.

### Importing libraries

First, we need to import necessary libraries. For our ML model, we will be using our own LinearRegression implemenation.

```python
import pandas as pd
import importlib
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
import pickle
```

```python
linear_regression = importlib.import_module('linear_regression')
```

### Data handling

Next, we need to import our input data file, which has already been preprocessed beforehand.

```python
# Load the data file

data = pd.read_csv('../output/data/real_estate_preprocessed.csv')
data = data[['district', 'size', 'floor', 'registration', 'rooms', 'parking', 'balcony', 'state', 'price']]
```

```python
# Split data into feature vectors and outputs

X = np.array(data.iloc[:, 0:-1])
y = np.array(data.iloc[:, -1:])
```

```python
# Use OneHotEncoder for categorical features

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse=False), [0, 7])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# Split data into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X.tolist(), y, test_size=0.2, random_state=0)
X_train = csr_matrix(X_train).toarray()
X_test = csr_matrix(X_test).toarray()
```

```python
# Use StandardScaler to scale train and test inputs

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Model training and evaluation

We will be using our own implemenation of Linear Regression. As our evaluation metric, we will be looking at both the RMSE (Root Mean Squared Error) and R2 (R squared) score.

```python
# Use self implemented LinearRegression regressor to fit train set and predict test set

regressor = linear_regression.LinearRegression(alpha=0.01, n_epoch=1000)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
# Visualizing RMSE by iteration

plt.plot(regressor.rmse)
plt.title('RMSE by iteration visualization')
plt.xlabel('Iteration')
plt.ylabel('RMSE')
plt.show()
```

```python
# Print RMSE

mean_squared_error(y_test, y_pred, squared=False)
```

```python
# Print R squared

r2_score(y_test, y_pred)
```

### Saving model, One Hot Encoder and Scaler

We use pickle to save our model, encoder and scaler, which we will need in our application for further predictions.

```python
with open('../output/linear_regression/linear_regression.pickle', 'wb') as out:
    pickle.dump(regressor, out)

with open('../output/linear_regression/ohe.pickle', 'wb') as out:
    pickle.dump(ct, out)

with open('../output/linear_regression/scaler.pickle', 'wb') as out:
    pickle.dump(scaler, out)
```

```python
with open("../output/linear_regression/coeff.txt", 'w') as output:
    for c in regressor.coeff:
        output.write(str(c) + '\n')
```
