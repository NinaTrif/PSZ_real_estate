### Function definitions

```python
def print_query_results(query):
    t = PrettyTable()

    cursor = conn.cursor()
    cursor.execute(query)
    t.field_names = cursor.column_names
    data = cursor.fetchall()
    for d in data:
        t.add_row(d)
    print(t)
```
